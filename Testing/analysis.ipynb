```python
def getPreds(df, method):
    if method == 0:
        df['BJP prediction'] = df['BJP Total count'] / (df['BJP Total count'] + df['INC Total count'] + df['AAP Total count'])
        df['INC prediction'] = df['INC Total count'] / (df['BJP Total count'] + df['INC Total count'] + df['AAP Total count'])
        df['AAP prediction'] = df['AAP Total count'] / (df['BJP Total count'] + df['INC Total count'] + df['AAP Total count'])
    elif method == 1:
        df['BJP prediction'] = df['BJP Positive count']/(df['INC Positive count'] + df['BJP Positive count'] + df['AAP Positive count'])
        df['INC prediction'] = df['INC Positive count']/(df['INC Positive count'] + df['BJP Positive count'] + df['AAP Positive count'])
        df['AAP prediction'] = df['AAP Positive count']/(df['INC Positive count'] + df['BJP Positive count'] + df['AAP Positive count'])
        df['BJP prediction'] = df['BJP Positive count']/(df['BJP Positive count'] + df['INC Positive count'] + df['AAP Positive count'])
        df['INC prediction'] = df['INC Positive count']/(df['BJP Positive count'] + df['INC Positive count'] + df['AAP Positive count'])
        df['AAP prediction'] = df['AAP Positive count']/(df['BJP Positive count'] + df['INC Positive count'] + df['AAP Positive count'])
    elif method == 2:
        df['BJP prediction'] = (df['INC Negative count']+df['AAP Negative count'])/(df['INC Negative count'] + df['BJP Negative count'] + df['AAP Negative count'])
        df['INC prediction'] = (df['BJP Negative count']+df['AAP Negative count'])/(df['INC Negative count'] + df['BJP Negative count'] + df['AAP Negative count'])
        df['AAP prediction'] = (df['INC Negative count'] + df['BJP Negative count'])/(df['INC Negative count'] + df['BJP Negative count'] + df['AAP Negative count'])
        df['BJP prediction'] = df['BJP prediction']/(df['BJP prediction'] + df['INC prediction'] + df['AAP prediction'])
        df['INC prediction'] = df['INC prediction']/(df['BJP prediction'] + df['INC prediction'] + df['AAP prediction'])
        df['AAP prediction'] = df['AAP prediction']/(df['BJP prediction'] + df['INC prediction'] + df['AAP prediction'])
    elif method == 3:
        BJP_pos = df['BJP Positive count']/(df['BJP Total count'])
        INC_pos = df['INC Positive count']/(df['INC Total count'])
        AAP_pos = df['AAP Positive count']/(df['AAP Total count'])
        df['INC prediction'] = INC_pos/(BJP_pos + INC_pos + AAP_pos)
        df['BJP prediction'] = BJP_pos/(BJP_pos + INC_pos + AAP_pos)
        df['AAP prediction'] = AAP_pos/(BJP_pos + INC_pos + AAP_pos)
    elif method == 4:
        BJP_neg = df['BJP Negative count']/(df['BJP Total count'])
        INC_neg = df['INC Negative count']/(df['INC Total count'])
        AAP_neg = df['AAP Negative count']/(df['AAP Total count'])
        df['INC prediction'] = (BJP_neg + AAP_neg)/(INC_neg + BJP_neg + AAP_neg) * 0.5
        df['BJP prediction'] = (INC_neg + AAP_neg)/(INC_neg + BJP_neg + AAP_neg) * 0.5
        df['AAP prediction'] = (BJP_neg + INC_neg)/(INC_neg + BJP_neg + AAP_neg) * 0.5
    elif method == 5:
        BJP_pos = abs(df['BJP Positive Sum'])
        INC_pos = abs(df['INC Positive Sum'])
        AAP_pos = abs(df['AAP Positive Sum'])
        df['INC prediction'] = INC_pos/(BJP_pos + INC_pos + AAP_pos)
        df['BJP prediction'] = BJP_pos/(BJP_pos + INC_pos + AAP_pos)
        df['AAP prediction'] = AAP_pos/(BJP_pos + INC_pos + AAP_pos)
    elif method == 6:
        BJP_pos = abs(df['BJP Positive Sum']/(df['BJP Total Sum']))
        INC_pos = abs(df['INC Positive Sum']/(df['INC Total Sum']))
        AAP_pos = abs(df['AAP Positive Sum']/(df['AAP Total Sum']))
        df['INC prediction'] = INC_pos/(BJP_pos + INC_pos + AAP_pos)
        df['BJP prediction'] = BJP_pos/(BJP_pos + INC_pos + AAP_pos)
        df['AAP prediction'] = AAP_pos/(BJP_pos + INC_pos + AAP_pos)
    return df
```

```python
dfs = [vader,bnb,vader_bnb,vader_subj,bnb_subj,vader]
methods = [0,1,2,3,4,5,6]
```

```python
senti_methods = {"vader":vader,"bnb":bnb,"vader_subj":vader_subj,"bnb_subj":bnb_subj,"vader_bnb":vader_bnb}
```

```python
res_dict = {
    "sentiment_method": [],
    "prediction_method": [],
    "BJP Prediction": [],
    "INC Prediction": [],
    "AAP Prediction": []
}
for senti in senti_methods:
    for method in methods:
        if senti == "bnb" and (method == 5 or method == 6):
            continue
        res = getPreds(get_sentiments(senti), method)
        res_dict["sentiment_method"].append(senti)
        res_dict["prediction_method"].append(method)
        res_dict["BJP Prediction"].append(res['BJP prediction'].mean())
        res_dict["INC Prediction"].append(res['INC prediction'].mean())
        res_dict["AAP Prediction"].append(res['AAP prediction'].mean())
```

```python
pd.DataFrame(res_dict).to_csv("Punjab_predictions.csv",index=False)
```
